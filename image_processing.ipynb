```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
# print current version 
print('cv version is {}'.format(cv2.__version__))
```

```python
# read image using cv.imread
cv_img = cv2.imread('images/butterfly.jpg')  # cv image is BGR format 
# convert to RGB format for display
rgb_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB)
# show image using matplotlib
plt.title('RGB image display using matplotlib')
plt.imshow(rgb_img)
```

```python
## color space transformation 
# convert to gray
gray_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)
# gray image display, using cmap='gray' properties
plt.title('display gray image')
plt.imshow(gray_img, cmap='gray')
print(gray_img.shape)
```

```python
# convert to HSV image 
hsv_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2HSV)
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(hsv_img[:,:,i], cmap='gray')
```

```python
## linear filter

# average filter
avg_img = cv2.blur(cv_img,(20,20))
plt.subplot(1,2,1); plt.title('original image') ;plt.imshow(cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2); plt.title('blurred image') ;plt.imshow(cv2.cvtColor(avg_img, cv2.COLOR_BGR2RGB))
```

```python
# gaussian filter
guass_img = cv2.GaussianBlur(cv_img,(21,21),5)
plt.subplot(1,2,1); plt.title('original image') ;plt.imshow(cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2); plt.title('guassian blurred image') ;plt.imshow(cv2.cvtColor(guass_img, cv2.COLOR_BGR2RGB))
```

```python
## non-linear filter 

# median filter
salt_img = cv2.imread('images/salt_noise.jpg')  # cv image is BGR format 
med_img = cv2.medianBlur(salt_img,5)

plt.subplot(1,2,1); plt.title('salt noise image') ;plt.imshow(cv2.cvtColor(salt_img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2); plt.title('median filterred image') ;plt.imshow(cv2.cvtColor(med_img, cv2.COLOR_BGR2RGB))
```
